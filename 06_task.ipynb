```python
df["Base Price"].fillna(0,inplace = True)
df["COST IN ₹ (CR.)"].fillna(0,inplace = True)
df["Cost IN $ (000)"].fillna(0,inplace = True)
df["2022 Squad"].fillna("NULL",inplace = True)
df["Base Price IN ₹"].fillna(0,inplace = True)
df["Base Price IN $"].fillna(0,inplace = True)
df
```

```python
Questions_from_ipl_2023_auction = ["1: Which teams spent the most on player acquisitions?",
                                   "2: What is the distribution of player types across Each Teams?",
                                   "3: How much did each team spend on different types of players (Batsman, Bowler, etc.)?",
                                   "4: Which team got the best value for money considering the type and cost of players?",
                                   "5: Who are the top 10 most expensive players in the 2023 season and which team purchased them?"] 
Questions_from_ipl_2023_auction 
```

```python
# First Question: Which teams spent the most on player acquisitions?

team_spending = df.groupby('Team')['COST IN ₹ (CR.)'].sum().sort_values(ascending = False)

plt.figure(figsize = (10,6))
ax = team_spending.plot(kind = 'bar' , color = 'skyblue')
for label in ax.containers:
     ax.bar_label(label)
plt.title("Total spending by Each Team (in ₹ (CR.))")
plt.xlabel("Teams Name")
plt.ylabel("Total spending (in ₹ (CR.) ")
plt.show()
```

```python
# Conclusion:

# Sunrisers Hyderabad spent most money in a auction ₹35.7Cr in year 2023.

# And Kolkata Knight Riders spent least amount in auction ₹5.4Cr in year 2023.
```

```python
# Second Question:  What is the distribution of player types across Each Teams?

# Brief :-How are different types of players (e.g., Batsman, Bowler, Wicketkeeper) distributed across the various IPL teams?

df1=pd.crosstab(df.Team,df.TYPE)
df1['Total Purchased Players'] = df1.sum(axis=1)
df1
```

```python
plt.figure(figsize = (10,5))
ax = sns.countplot(x = "Team", data=df, hue = "TYPE")
for label in ax.containers:
    ax.bar_label(label)
plt.title("Type of players purchased by each Team")
plt.xlabel("Teams Name")
plt.ylabel("No. of Players")
plt.xticks(rotation=90)
plt.show()
```

```python
# As we can see many players got unsold.

# After observing above chart it is easily visible every teams mainly focus was on 'ALL ROUNDERS'.

# After 'ALL ROUNDERS' every teams gave their focus on their bowling lineup which is very importnant to win tournament.
```

```python
# Third Question: How much did each team spend on different types of players (Batsman, Bowler, etc.)?

df2 = pd.crosstab(index=df['Team'],columns=df['TYPE'],values=df['COST IN ₹ (CR.)'],aggfunc='sum')
df2['Total Spent'] = df2.sum(axis=1)
df2
```

```python
team_spending_by_type = df.groupby(['Team','TYPE'])['COST IN ₹ (CR.)'].sum().unstack()

team_spending_by_type.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Set3')
plt.title('Team Spending by Player Type (in ₹ Cr.)')
plt.xlabel('Team')
plt.ylabel('Total Spending (₹ Cr.)')
plt.xticks(rotation=45)
plt.legend(title='Player Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Fourth Question: Which team got the best value for money considering the type and cost of players?

team_spending = df.groupby(["Team","TYPE"])["COST IN ₹ (CR.)"].mean().unstack().fillna(0)

plt.figure(figsize = (12,8))
sns.heatmap(team_spending,cmap='GnBu',annot=True,linewidths=0.5,fmt='.2f')

plt.title('Average Cost per Player Type by Team (in ₹ Cr.)')
plt.xlabel('Player Type')
plt.ylabel('Team')
plt.show()
```

```python
# Fifth Question: Who are the top 10 most expensive players in the 2023 season and which team purchased them?

expensive_player = df.sort_values(by='COST IN ₹ (CR.)', ascending=False).head(10)

plt.figure(figsize = (10,6))
plt.barh(expensive_player['Player'],expensive_player['COST IN ₹ (CR.)'],color = 'teal')
plt.gca().invert_yaxis()  # Invert y-axis to show the most expensive player on top

plt.xlabel('Cost in ₹ (CR.)')
plt.ylabel('Player')
plt.title('Top 10 Most Expensive Players in IPL 2023')
plt.show()
```
